The activation function decides whether a neuron should be activated or not by calculating the weighted sum and further adding bias to it. The purpose of the activation function is to introduce non-linearity into the output of a neuron the most important thing is that activation functions introduce non-linearity into the network. A neural network without activation functions is basically just a linear regression model and is not able to do more complicated tasks such as language translations and image classifications. Also, linear functions are not able to make use of backpropagation

During the calculations of the values for activations in each layer, we use an activation function right before deciding what exactly the activation value should be. From the previous activations, weights & biases in each layer, we calculate a value for every activation in the next layer. But before sending that value to the activations of the next layer, we use an activation function to scale the output

### 1. Sigmoid Function
Sigmoid is a non-linear activation function used mostly in feedforward neural networks. It is a bounded differentiable real function, defined for real input values.
- bounded : 범위가 제한 (0 ~ 1 ) → binary classification 마지막 단게에서 '확률' 을 나타낼 때 자주 쓰임
- differentiable  : 미분가능, 모든 지점에서 기울기를 구할 수 있음
- Defined for real input values : 어떤 실수($-\infty$ ~ $+\infty$)를 넣어도 결과값

---
- Vanishing gradient

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

a = np.array([-20, -1.0, 0.0, 1.0, 20])
b = sigmoid(a)
print(b)
```

#### Hard sigmoid
- A hard sigmoid is a non-smooth function used in place of a sigmoid function. (꺾이는 지점)
- These retain the basic shape of a sigmoid, rising from 0 to 1, but using simpler functions(복잡한 지수함수 대신에 min, max 같은 단순 비교 연산이나 일차 함수 (ax + b) 조합에서 시그모이드와 비슷한 흉내)
- and offer less loss function (지수함수에 비해 단순화된 함수 사용함 → 연산 속도가 빨라 연산 능력에 낮은 모바일 기기나 임베디드시스템에서 효율적)

```python
def Hard_Sigmoid(x):
  return np.clip((x + 1) / 2, 0, 1) #..?
```

```python
a = np.array([-20, -1.0, 0.0, 1.0, 20])
b = Hard_Sigmoid(a)
print(b)
```

### 2. Hyperbolic Tangent Function (Tanh)
The hyperbolic tangent function is another type of AF used in DL and it has some variants used in DL applications. The hyperbolic tangent function is known as tanh function, whose range lies between -1 to 1, thus the output of the tanh function is given by

f (x) = ( e^ x − e^ −x/ e^ x + e^ −x )

The tanh function became the preferred function compared to the sigmoid function in that it gives better training performance for multi-layer neural networks. However, the tanh function could not solve the vanishing gradient problem

Usually used in hidden layers of a neural network

- -1 ~ 1
- 출력이 0을 중심으로 형성(Zero-centered)되기 때문에, 다음 층으로 전달되는 데이터가 편향되지 않아 학습이 시그모이드보다 더 빠르고 안정적으로 진행됩니다.
- Vanishing Gradient : 시그모이드와 마찬가지로, 입력값($x$)이 너무 크거나 작아지면 그래프가 평평해집니다. 이때 미분값이 0에 가까워지면서 학습 신호가 끊기는 문제는 여전합니다. (다만, 시그모이드보다는 기울기가 가팔라서 조금 덜하긴 합니다.)
- 보통 신경망의 **은닉층(Hidden Layers)**에서 사용됩니다.

---
- still vanishing gradient

```python
def tanh(x):
  return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
a = np.array([-20, -1.0, 0.0, 1.0, 20])
b = tanh(a)
print(b)
```

### 3. Softmax
- The softmax function is also a type of sigmoid non-linear function but is handy when we are trying to handle classification problems. (sig - binary / softmax - multi)
- Usually used when trying to handle multiple classes. The softmax function would squeeze the outputs for each class between 0 and 1 and would also divide by the sum of the outputs.
소프트맥스 함수는 각 클래스에 대한 출력값을 0과 1 사이로 압축하고, 이를 전체 출력값의 합으로 나눕니다.

- The softmax function is ideally used in the output layer
모델의 최종 결론을 "확률" 형태로 내놓아야 하기 때문에 마지막 단계에 위치합니다.

f (xi) = ( exp (xi) /sum( exp( xj)))

만약 인공지능이 사진을 보고 [개, 고양이, 토끼] 세 가지 중 하나로 분류한다고 가정해 봅시다.
1. Raw Scores (Logits): 모델이 계산한 점수가 [2.0, 1.0, 0.1]이라고 합시다.
2. Exp 적용: 각 점수에 지수를 취합니다. [7.4, 2.7, 1.1] (값이 커지고 모두 양수가 됨)
3. Sum 계산: 모두 더합니다. 7.4 + 2.7 + 1.1 = 11.2
4. Softmax 결과: 각 값을 합으로 나눕니다.
- 개: $7.4 / 11.2 = \mathbf{0.66}$ (66%)
- 고양이: $2.7 / 11.2 = \mathbf{0.24}$ (24%)
- 토끼: $1.1 / 11.2 = \mathbf{0.10}$ (10%)
- 합계: $0.66 + 0.24 + 0.10 = \mathbf{1.0}$ (100%)



```python
def softmax(x):
    e_x = np.exp(x - max(x))
    return e_x / e_x.sum()

a = np.array([-20, -1.0, 0.0, 1.0, 20])
b = softmax(a)
print(b)
```

### 4. Relu
Stands for Rectified linear unit. It is the most widely used non-linear activation function. Chiefly implemented in hidden layers of Neural network.

f (x) = max (0, x)

The main advantage of using the rectified linear units in computation is that, they guarantee faster computation

ReLu is less computationally expensive than tanh and sigmoid because it involves simpler mathematical operations

In simple words, RELU learns much faster than sigmoid and Tanh function.

ReLU units can be fragile during training and can "die". For example, a large gradient flowing through a ReLU neuron could cause the weights to update in such a way that the neuron will never activate on any datapoint again. If this happens, then the gradient flowing through the unit will forever be zero from that point o

```python
import tensorflow as tf
import numpy as np

# 1. 직접 함수를 정의할 때 (텐서 연산이 가능하도록 수정)
def relu(x):
    return tf.maximum(0.0, x)

foo = tf.constant([-10, -5, 0.0, 5, 10], dtype=tf.float32)

# 2. 직접 만든 함수 사용 시
y_custom = relu(foo).numpy()

# 3. 케라스 내장 함수 사용 시 (작성하신 코드의 올바른 실행)
y_keras = tf.keras.activations.relu(foo).numpy()

print(f"입력값: {foo.numpy()}")
print(f"결과값: {y_keras}")
```
